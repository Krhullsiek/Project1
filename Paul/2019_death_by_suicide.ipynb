```python
import pandas as pd
```

```python
#Import CSV File (Downloaded from CDC Wonder API - https://wonder.cdc.gov/ucd-icd10.html)

cdc_file = 'Underlying Cause of Death, 1999-2019-2.csv'
mortality_data = pd.read_csv(cdc_file)
```

```python
# Drop last rows and empty 'Notes' column - last rows are search metadata from CDC
mortality_data.drop(mortality_data.tail(88).index, inplace = True)
mortality_data.drop('Notes', axis=1, inplace=True)

#format 'County Code' as int.
mortality_data['County Code'] = mortality_data['County Code'].astype(int)
mortality_data.head(20)
```

```python
test_df = mortality_data
```

```python
test_df['Age Adjusted Rate'] = pd.to_numeric(test_df['Age Adjusted Rate'], errors='coerce', downcast="float")
test_df.dtypes
```

```python
test_df['Crude Rate'] = pd.to_numeric(test_df['Crude Rate'], errors='coerce', downcast="float")
test_df.dtypes
```

```python
test_df
```

```python
m_group = test_df.groupby('County Code')

test_df['Deaths sum'] = m_group['Deaths'].transform(lambda x: x.sum())
test_df
#lgst_cities = city_grp.apply(lambda g: g[g['Population'] == g['Population'].max()])
```

```python
test_df['summed_crude_rate'] = (test_df['Deaths sum']/(test_df['Population']/100000))
```

```python
# test_df['summed_adjusted_rate'] = (test_df['summed_crude_rate'] * (test_df['Age Adjusted Rate'] * ((test_df['Population']/100000) / test_df['Deaths sum'])))

test_df['summed_adjusted_rate'] = (test_df['summed_crude_rate'] * (test_df['Age Adjusted Rate'] / test_df['Crude Rate']))
```

```python
test_df.head(40)
```

```python
clean_df = test_df.drop_duplicates(subset=['County Code'], keep='last')
clean_df
```

```python
clean_df.head(40)
```

```python
my_df = clean_df[['County', 'County Code', 'Deaths sum', 'summed_crude_rate', 'summed_adjusted_rate']]
my_df
```

```python
my_df['County Code']= my_df['County Code'].apply(str)
my_df['County Code']= my_df['County Code'].str.zfill(5)
my_df
```

```python
my_df.to_csv('county_suicide_rates.csv')
```
