```python
import pandas as pd
import requests
from pprint import pprint
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
df = pd.read_csv ('../cy/large_cities.csv')
df
```

```python
percentiles_df = pd.DataFrame()
index = 0
for ind in df.index:
    counter = 0
    city = df['city'][ind].replace(' ', '-')
    city = city.lower()
    
    url = (f'https://api.teleport.org/api/urban_areas/slug:{city}/salaries/')
    
    response = requests.get(url).json()
    json = response

    try:
            percentile_25 = 0
            percentile_50 = 0
            percentile_75 = 0
            for job in json['salaries']:
                percentile_25 += (json['salaries'][counter]['salary_percentiles']['percentile_25'])
                percentile_50 +=(json['salaries'][counter]['salary_percentiles']['percentile_50'])
                percentile_75+=(json['salaries'][counter]['salary_percentiles']['percentile_75'])
                counter+=1
            
    except:
        KeyError
    try:    
        df.loc[ind,'city'] =city
        low_percentile = percentile_25 / (len(json['salaries']))
        mid_percentile = (percentile_50  / (len(json['salaries'])))
        high_percentile = (percentile_75 / (len(json['salaries'])))
        mean = (low_percentile + mid_percentile + high_percentile) / 3
        df.loc[ind,'25th_percentile'] = low_percentile
        df.loc[ind,'50th_percentile'] = mid_percentile
        df.loc[ind,'75th_percentile'] = high_percentile
        df.loc[ind,'mean'] = mean
        
        index+=1

    except:
        KeyError
df
```

```python
percentiles_df = df.dropna()
diff = percentiles_df['75th_percentile'] - percentiles_df['25th_percentile']
plt.xticks(rotation = 90)
plt.bar(percentiles_df['city'], diff)
plt.title('Difference b/t 75th and 25 percentiles of income in dollars')
```

```python
plt.bar(percentiles_df['city'],percentiles_df['mean'])
plt.xticks(rotation = 90)
plt.show()
```
